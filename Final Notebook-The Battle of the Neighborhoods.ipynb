### 3.2 Retrieve the data
#### 3.2.1 Wikipedia data 

Use the BeautifulSoup library to extract the page from Wikipedia and then parse the page using python. The list is an unordered html list of which the page contains more than one.  First use BeautifulSoup's FindAll method to get all of the unordered lists, and then use a loop to determine which html "\<ul\>" tag contains "Arbutus-Greenway" and is therefore our target. 

```python
# build the URL
url='https://en.wikipedia.org/wiki/List_of_neighbourhoods_in_Vancouver'

# make a GET request to retrieve the page
wikidata = requests.get(url).text

# create a beautiful soup object containing the page data
soup = BeautifulSoup(wikidata,"html5lib")

# create an empty dictionary
neighborhoods = [] # initially empty list

#find un-ordered lists
ul_list = soup.findAll('ul')

ul_index = -1       # catch-all incase we don't find any <ul> tags
for index,ul in enumerate(ul_list):
    if('Arbutus_Greenway') in str(ul):
        ul_index = index
if (ul_index == -1):
    print("list not found")
else:
    print("the index of the neighborhood ul element is ", ul_index)
```

The unordered list we want is the one with index 1.  <br/>
Each neighborhood is identified by an anchor tag "\<a href=...\>" within each list item ("\<li\>") and the description contains a repeat of the neighborhood name.  Parse the rows of the ul_list soup object to get the name and description and use python to strip out the redundant name.

```python
#Get the desired unordered list row from the Result Set at our index of interest.
li_list = ul_list[ul_index]
 
for li in li_list.findAll('li'):              # find the <li> tags
    cell = {}                                 # empty dictioinary
                                              # now loop inside the <li> item
    for link in li.find('a'):                 # find the first anchor tag (<a>)
        cell['Name'] = link.lstrip().rstrip() # assign the dictionary Name value and remove white space before and after the name
    
    basedesc = str(li.text)                   # get the text of the list item and strip out the duplicate name and hyphen
    strindex = basedesc.find('-')
    if (strindex > 0):
        strstart = strindex + 1 
    strend = len(basedesc) - strstart
                                               # assign the dictionary Description value and remove any leading white space
    cell['Description']=basedesc[strstart:strend].lstrip()
    
    neighborhoods.append(cell)                 # add the dictionary item to the neighborhoods list
       
```

Finally, create a Pandas DataFrame of the neighborhoods list.  One of the neighborhoods needs to be renamed to match its official name so that other datasets can be merged with this one.

```python
neighborhoodsdf = pd.DataFrame(neighborhoods)               # create a Pandas Datafram

neighborhoodsdf.replace("Arbutus Ridge",                    # clean up the name of this neighborhood to match its official name
           "Arbutus-Ridge",                    
           inplace=True)

print ("Neighborhoodsdf DataFrame created!")
```

```python
# get dog parks within 1.5km of the neighborhood's centroid

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

Create a function to get any existing Dog Parks in each neighborhood. The function will be called once for each neighorhood

```python
# a function that returns the dog parks for all neighborhoods
def getNearbyDogParks(hoodnames, lats, lons, radius):
 

    search_query='Dog Park'
    parks_list=[]
    

    for hoodname,lat,lon in zip(hoodnames,lats,lons):
        parkurl = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&oauth_token={}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID
                       , CLIENT_SECRET, lat, lon,ACCESS_TOKEN, VERSION, search_query, radius, LIMIT)
    
# call the GET request 
        results = requests.get(parkurl).json()['response']['venues']  
    
        parks_list.append([(hoodname, 
                        lat, 
                        lon,
                        p['name'], 
                        get_category_type(p),
                        p['location']['lat'],
                        p['location']['lng'] ,
                        p['location']['distance']) for p in results])

    nearbyDogParks = pd.DataFrame([park for park_list in parks_list for park in park_list])
    nearbyDogParks.columns = ['Neighborhood',
                              'Neighborhood_Latitude',
                              'Neighborhood_longitude',
                              'DogPark',
                              'Category',
                              'DogPark_Latitude',
                              'DogPark_Longitude',
                              'DogPark_Distance']
    
    return nearbyDogParks
```

### 3.4 Clustering Analysis

There are several algorithms for use in cluster analysis - for example, density-based clustering, hierarchical clustering, fuzzy clustering and k-means.  For this project, we will use the K-Means clustering algorithm, which is a  commonly used technique, is one of the simplest and works well with unlabelled data. With K-Means analysis, we will identify clusters of neighborhoods based on the number of dogs per hectare (dog density), the number of dog parks per hectare (dog park density) and number of pubs per hectare (pub density). .


#### 3.4.1 Preparation
The K-Means clustering algorithm requires that the data are continuous (not categorical), so we will remove columns that are not continuous.  Then we will use StandardScaler, a popular method for normalization of data that results in a mean and standard deviation of zero. 


```python
#Reduce complexity by removing unnecessary, categorical and redundant variables
kmdata = dogsAndSuds.drop(columns=['Latitude','Longitude','ShortName','Neighborhood','Description','Polygon', 'Area','Population','LicensedDogs','ParkCount','PubCount'])
kmdata
```

#### 3.4.2 Normalize the data using the StandardScaler method

The k-Means algorithm along with many other Machine Learning algorithms requires that all data are continuous variables and that the data have been normalized, and it is obvious from the data that some of the columns have drastically different scale form others. To mitigate against this we will use the StandardScaler which is a common normalization method and leaves the data with zero mean and standard deviation.

```python
from sklearn.preprocessing import StandardScaler
X = kmdata.values[:,:] #[:,1]
kmdata_norm = StandardScaler().fit_transform(X)
kmdata_norm
```

#### 3.4.3 Determine the optimum value of K

To determine how many clusters (K), it is common to run the k-means algorithm several times and evaluate the impact of changing 'k' on the within cluster sum of squares (squared Euclidean distance from the value to the mean), also called inertia, each time. 
We will use the 'Elbow' method for this evaluation.

```python
# create a list of sum of squared errors (distance from the mean**2)
sse = []

Krange = range(1,15)
for k in Krange:
    #train the model with the current value of k
    kmeans = KMeans(n_clusters = k)
    kmeans = kmeans.fit(kmdata_norm)
    
    #save the resultant sum of squares error
    sse.append(kmeans.inertia_)
sse   
```

#### 3.4.4 Plot Inertia by the number of clusters requested

```python
plt.plot(Krange, sse, 'bx-')
plt.title('Figure 8 - Evaluation of the number of clusters on Inertia (Elbow method)')
plt.xlabel('# of clusters (k)')
plt.ylabel('Inertia')

plt.show()
```

##### The inflection point (the elbow) at 4 clusters suggests that this is the optimal value for K. 

```python
kOpt=4
#train the model
k_means = KMeans(init = "k-means++", n_clusters = kOpt, n_init = 10)
k_means.fit(kmdata_norm)
clusternum = k_means.labels_
kmdata['Group'] = clusternum+1
```

#### 3.4.5 Plot the clusters

```python
figure(num=None, figsize=(12, 9), dpi=80, facecolor='w', edgecolor='k')
plt.scatter(kmdata.iloc[:, 1], 
            kmdata.iloc[:, 0], 
            c=kmdata.iloc[:,3], 
            s=200, 
            cmap='rainbow')
centers = kmeans.cluster_centers_
plt.title("Figure 9 - 2D scatter plot of neighborhood clusters")
plt.show()
```

The two-dimensional plot clearly shows that the neighborhoods can be separated by the selected features dog density, pub density and dogpark density.

The recent COVID-19 epidemic may also have played a role in the closure of some establishments that nevertheless still appear in the FourSquare data.  Further analysis may tease out those instances. <br/> <br/>
    We may also have underestimated dog park density in some neighborhoods since the FourSquare results only included venues in which the name contains "Dog Park" or the FourSquare venue had category "Dog Run".  However, we see no <i>a priori</i> reason that this underestimate would not be uniformly distributed across the neighborhoods.  <br/><br/>
    Further analysis could focus on other variables such as socio-economic status of the neighborhoods, a more detailed classification of existing pubs (for example, are some existing pubs 'dog friendly'?) or a broader analysis of dog parks to include other types of parks that dog owners frequent with their pet. Given the recommendations provided here, more detailed breakdown of these two neighborhoods by city block could be used to identify the actual addresses most likely to be successful. <br/> <br/>
    Finally, location is only one in a large number of factors that will influence establishment of a dog bar.  Subsequent studies may focus on economic feasibility given the cost of property across neighborhoods, licensing and other factors.

## Conclusion

Dog owners love their pets and need to take them out for exercise.  Dog bars allow them to do so while consuming their favorite beverage in a dog-friendly environemt, and in doing so enjoy social interaction with like-minded pub-going dog owners.  In this study we have gathered and analyzed data from several sources to identify two neighborhoods in which to locate Vancouver's first dog bar.  We conclude that the neighborhoods of Grandview-Woodlands and Kitsilano provide the best mix of a dog-owning population, existing pubs to attract visitors and that are poorly served by existing dog parks.
